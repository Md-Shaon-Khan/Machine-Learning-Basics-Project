```python
import numpy as np
import pandas as pd
```

```python
breast = pd.read_csv('breast_cancer_dataset.csv')
breast.head()
```

```python
breast.shape
```

```python
breast.info()
```

```python
breast.describe()
```

```python
breast = breast.drop(['id','Unnamed: 32'],axis='columns')
breast
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
breast['diagnosis'] = label.fit_transform(breast['diagnosis'])

breast.head(5)
```

```python
breast.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(30,30))
sns.heatmap(breast.corr(),cmap='coolwarm',annot=True)
plt.show()
```

```python
corr=breast.corr()
corr_target = corr['diagnosis'].sort_values(ascending=False)
corr_target
```

```python
sns.countplot(x=breast['diagnosis'])
plt.title('Distribution of Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
top_features = corr[1:21].index.tolist()
```

```python
plt.figure(figsize=(20,25))
for i,feature in enumerate(top_features):
    plt.subplot(5,4,i+1)
    sns.boxplot(x='diagnosis', y=feature, data=breast, palette='coolwarm')
    plt.title(f'{feature} vs Diagnosis', fontsize=12)
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 25))
for i, feature in enumerate(top_features):
    plt.subplot(5, 4, i + 1)
    sns.stripplot(x='diagnosis', y=feature, data=breast, palette='coolwarm', alpha=0.6, jitter=0.25)
    plt.title(f'{feature} vs Diagnosis', fontsize=11)
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = breast[top_features]
y = breast['diagnosis']

X_train, X_test ,y_train, y_test = train_test_split(X,y,test_size=0.23,random_state=42)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```
